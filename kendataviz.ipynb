```python
import pandas as pd
import os
from functools import reduce
import matplotlib.pyplot as plt

# Using year 2000 as the reference year

df = pd.read_csv("merged_nuclear_data.csv")
df  = df[(df['Year'] >= 2000) & (df['Year'] <= 2025)]
print(df.head())
```

```python
df.plot(x='Year',y=['nuclear_support_pct', 'nuclear_opposition_pct'],kind='line',marker='o')
plt.title('Overall, do you strongly favor, somewhat favor, somewhat oppose or strongly oppose\n'
'the use of nuclear energy as one of the ways to provide electricity for the U.S.?')
plt.xlabel('Year')
plt.ylabel('Percent')
plt.legend(['Support Nuclear', 'Oppose Nuclear'])
plt.show()
```

```python
cols = ['nuclear_support_pct','nuclear_opposition_pct']
df_plot = df.dropna(subset=['Year']).dropna(subset=cols, how='all')
df_plot.plot(x='Year',y=cols,kind='line')
plt.title('Overall, do you strongly favor, somewhat favor, somewhat oppose or strongly oppose\n'
'the use of nuclear energy as one of the ways to provide electricity for the U.S.?')
plt.xlabel('Year')
plt.ylabel('Percent')
plt.legend(['Support Nuclear', 'Oppose Nuclear'])
plt.show()
```

```python
cols = ['republican_nuclear_support_pct','independents_nuclear_support_pct','democrats_nuclear_support_pct']

# converts number% to just numeric
for c in cols:
    df[c] = pd.to_numeric(df[c].astype(str).str.replace('%', '', regex=False), errors='coerce')

# keeps rows with a year and @ least one party value
df_plot = df.dropna(subset=['Year']).dropna(subset=cols, how='all')
df_plot.plot(x='Year',y=cols,kind='line',title='Overall, do you strongly favor, somewhat favor, somewhat oppose or strongly oppose the\n'
'the use of nuclear energy as one of the ways to provide electricity for the U.S.?')
plt.ylabel('Percent')
plt.legend(['Republicans', 'Independents', 'Democrats'])
plt.show()
```

```python
cols = ['pct_prefer_nonrenewable_energy','pct_prefer_alternative_energy']
df_plot = df.dropna(subset=['Year']).dropna(subset=cols, how='all')
df_plot.plot(x='Year',y=cols,kind='line',title='Which of the following approaches to solving the nations energy problems do you \n'
             'think the U.S. should follow right now  emphasize production of more oil, gas and coal \n'
             'supplies or emphasize the development of alternative energy such as wind and solar power?')
plt.ylabel('Percent')
plt.legend(['Republicans', 'Independents', 'Democrats'])
plt.show()
```

```python
df_plot = df.dropna(subset=['Year']).dropna(subset=cols, how='all')
df_plot.plot(x='Year',y='pct_concern_energy_affordability',kind='line',title='How much do you personally worry about the availability and affordability of energy?')
plt.ylabel('Percent')
plt.show()
```

```python
df_plot = df.dropna(subset=['Year']).dropna(subset=cols, how='all')
df_plot.plot(x='Year',y='nuclear_generation_gwh',kind='line',title='Nuclear Electricity Generation in the U.S.')
plt.ylabel('GWh')
plt.show()
```
